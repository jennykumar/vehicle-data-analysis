```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv"
df = pd.read_csv(url)

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.dropna()
```

```python
df['mpg'].hist()
plt.title("MPG Distribution")
plt.xlabel("Miles Per Gallon (MPG)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.scatter(df['horsepower'], df['mpg'])
plt.title("Horsepower vs MPG")
plt.xlabel("Horsepower")
plt.ylabel("MPG")
plt.show()
```

```python
df['origin'].value_counts().plot(kind='bar')
plt.title("Cars by Origin")
plt.xlabel("Origin")
plt.ylabel("Number of Vehicles")
plt.show()
```

Observation:
- MPG is negatively correlated with horsepower and weight
- Strong relationships exist between engine-related features

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.heatmap(df.select_dtypes(include=['number']).corr(), annot=True, cmap='coolwarm')
plt.title("Feature Correlation Heatmap")
plt.show()
```

📊 Insights

1. Vehicles with higher horsepower tend to have lower fuel efficiency.
2. Most vehicles are from a few dominant regions.
3. Fuel efficiency varies significantly across the dataset.
